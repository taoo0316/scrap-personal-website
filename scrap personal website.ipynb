```python
# import library
from bs4 import BeautifulSoup
import requests
# Request to website and download HTML contents
url='https://taoo0316.github.io/'
req=requests.get(url)
content=req.text
```

```python
soup=BeautifulSoup(content)
```

```python
print(content)
```

```python
print(soup)
```

```python
print(soup.prettify())
```

```python
for link in soup.find_all('a'):
    print(link.get('href'))
```
